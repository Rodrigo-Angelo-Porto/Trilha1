## Questão 2 📝

Faça uma calculadora simples. Peça ao usuário para digitar 2 números via `input()` e exiba o resultado das 4 operações básicas, cada uma em um `print()` separado.

````python
def calculadora():
  numero_1 = float(input("Escolha um número: "))
  numero_2 = float(input("Escolha outro número: "))
  soma = numero_1 + numero_2
  subtração = numero_1 - numero_2
  multiplicação = numero_1 * numero_2
  # Percebi depois de testes que ao dividir por 0 o código dava erro por não ser possível, então coloquei a condicional para evitar que ele divida e retorne ao usuário #
  if numero_2 != 0:
    divisão = numero_1 / numero_2
  else:
    divisão = "Não é possível dividir por 0"
  return soma, subtração, multiplicação, divisão
soma, subtração, multiplicação, divisão = calculadora()
print(soma)
print(subtração)
print(multiplicação)
print(divisão)
````

## Questão 6 📝

Crie um programa que recebe o salário de um funcionário e calcula o imposto devido com base nas faixas abaixo:

| Faixa salarial | Alíquota |
|---|---|
| Até R\$ 1.500 | Isento (0%) |
| R\$ 1.500,01 – R\$ 3.000 | 7,5% |
| R\$ 3.000,01 – R\$ 4.500 | 15% |
| Acima de R\$ 4.500 | 22,5% |

Exiba o salário bruto, o valor do imposto e o salário líquido.

````python
def salario_imposto():
  # Optei for usar esse replace para adequar tanto o input quanto o altput ao modo nacional de escrever valores, que retornam como é lido normalmente no brasil e o código não dá erro caso o usuário dê input de um número com vírgula, pois será trocado para ponto e o pyhton conseguirá realizar as operações #
  salario = float(input("Insira seu salário: ").replace(",", "."))
  if salario <= 1500:
    aliquota = 0
  elif salario > 1500 and salario <= 3000:
    aliquota = 0.075
  elif salario > 3000 and salario <= 4500:
    aliquota = 0.15
  else:
    aliquota = 0.225
  imposto = salario * aliquota
  liquido = salario - imposto
  salario_formato_brasileiro = f"{salario:.2f}".replace(".", ",")
  imposto_formato_brasileiro = f"{imposto:.2f}".replace(".", ",")
  liquido_formato_brasileiro = f"{liquido:.2f}".replace(".", ",")
  print(f"Salário bruto total: {salario_formato_brasileiro}")
  print(f"Imposto total a pagar: {imposto_formato_brasileiro}")
  print(f"Salário líquido total: {liquido_formato_brasileiro}")
  return
  # Optei por mesmo sem ser necessário colocar ":.2f", pois no contexto monetário sempre se coloca duas casas após a vírgula #
salario_imposto()
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use `if`/`elif`/`else` para cada faixa. Calcule `imposto = salario * aliquota`, depois `liquido = salario - imposto`. *

## Questão 7 📝

Crie um programa que verifica se um ano digitado pelo usuário é bissexto.

**Regras:**
- Divisível por 4 → bissexto
- Mas se divisível por 100 → **NÃO** é bissexto
- A menos que também seja divisível por 400 → **É** bissexto

Exiba `"Ano bissexto"` ou `"Não é bissexto"`.

````python
def ano_bissexto():
  ano = int(input("Digite um ano: "))
  if ano % 4 == 0 and (ano % 100 != 0 or ano % 400 == 0):
    resultado = "é bissexto"
  else:
    resultado = "não é bissexto"
  return ano, resultado
ano, resultado = ano_bissexto()
print(f"{ano} {resultado}")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use o operador `%` para verificar divisibilidade. Combine condições com `and` e `or`.*

## Questão 8 📝

Construa um programa que recebe o peso e a altura do usuário, calcula o IMC e classifica o resultado:

| IMC | Classificação |
|---|---|
| IMC < 18,5 | Abaixo do peso |
| 18,5 <= IMC < 25 | Peso normal |
| 25 <= IMC < 30 | Sobrepeso |
| IMC >= 30 | Obesidade |

Exiba o IMC com 2 casas decimais e a classificação.

````python
def classificar_imc():
  peso = float(input("Digite seu peso: ").replace(",", "."))
  altura = float(input("Digite sua altura: ").replace(",", "."))
  imc = peso/altura**2
  if imc < 18.5:
    resultado = "abaixo do peso"
  elif imc >= 18.5 and imc < 25:
    resultado = "com peso normal"
  elif imc >= 25 and imc < 30:
    resultado = "com sobrepeso"
  else:
    resultado = "com obesidade"
  peso_formato_brasileiro = f"{peso}".replace(".", ",")
  altura_formato_brasileiro = f"{altura}".replace(".", ",")
  print(f"Com base no seu peso, {peso_formato_brasileiro}Kg., e altura, {altura_formato_brasileiro}m., seu IMC é igual a {imc:.2f} e você está {resultado}")
  return
classificar_imc()
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*`IMC = peso / altura**2`. Para formatar com 2 casas decimais, use `f"{imc:.2f}"`.*

## Questão 9 📝

Peça ao usuário para digitar uma palavra e use métodos de string para:
1. Exibir com a primeira letra maiúscula
2. Exibir quantas letras ela tem
3. Exibir a palavra de trás para frente (slicing reverso)
4. Verificar se ela começa com a letra `"a"` (sem diferenciar maiúsculas de minúsculas)

````python
def alterar_palavra():
  palavra = input("Digite uma palavra: ")
  palavra_maiuscula = palavra.capitalize()
  print(f"Palavra com a primeira letra maiúscula: {palavra_maiuscula}")
  print(f"A palavra {palavra} tem {len(palavra)} letras")
  palavra_tras_frente = palavra[::-1]
  print(f"{palavra} de trás para frente: {palavra_tras_frente}")
  primeira_letra = palavra[0]
  if "A" in primeira_letra or "a" in primeira_letra:
    print(f"A palavra {palavra} começa com a letra 'A'.")
  else:
    print(f"A palavra {palavra} não começa com a letra 'A'")
  return
alterar_palavra()
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use `.capitalize()`, `len()`, slicing `[::-1]` e `.lower().startswith("a")`.*

## Questão 10 📝

Crie um classificador de triângulos. Receba os 3 lados pelo `input()` e:

1. Verifique se os lados formam um triângulo válido *(a soma de dois lados quaisquer deve ser maior que o terceiro)*
2. Se válido, classifique:
   - **Equilátero:** todos os lados iguais
   - **Isósceles:** dois lados iguais
   - **Escaleno:** todos diferentes

Exiba a classificação ou `"Não forma um triângulo"`.

````python
def classificação_triangulo():
    lado_1 = float(input("Digite o tamanho do lado 1: ").replace(",", "."))
    lado_2 = float(input("Digite o tamanho do lado 2: ").replace(",", "."))
    lado_3 = float(input("Digite o tamanho do lado 3: ").replace(",", "."))
    if lado_1 <= 0 or lado_2 <= 0 or lado_3 <= 0:
      print("Não se pode existir um lado de triângulo igual a 0 ou menos.")
      return
    lado_1_formato_brasileiro = f"{lado_1}".replace(".", ",")
    lado_2_formato_brasileiro = f"{lado_2}".replace(".", ",")
    lado_3_formato_brasileiro = f"{lado_3}".replace(".", ",")
    if lado_1 + lado_2 <= lado_3:
        print("Não forma um triângulo")
    elif lado_1 + lado_3 <= lado_2:
        print("Não forma um triângulo")
    elif lado_2 + lado_3 <= lado_1:
        print("Não forma um triângulo")
    else:
        if lado_1 == lado_2 == lado_3:
            print(f"O triângulo com as medidas {lado_1_formato_brasileiro}, {lado_2_formato_brasileiro} e {lado_3_formato_brasileiro} é um triângulo equilátero")
        elif lado_1 == lado_2 != lado_3 or lado_1 == lado_3 != lado_2 or lado_2 == lado_3 != lado_1:
            print(f"O triângulo com as medidas {lado_1_formato_brasileiro}, {lado_2_formato_brasileiro} e {lado_3_formato_brasileiro} é um triângulo isósceles")
        else:
            print(f"O triângulo com as medidas {lado_1_formato_brasileiro}, {lado_2_formato_brasileiro} e {lado_3_formato_brasileiro} é um triângulo escaleno")
    return
classificação_triangulo()
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Condição de triângulo válido: `a+b>c AND a+c>b AND b+c>a`. Depois compare os lados com `==` para classificar.*
